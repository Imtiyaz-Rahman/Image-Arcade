## Different Image API tests

This will show the extraction of images from the Art provided by the different APIs 

```python
# Run first for the tests
import requests
# Import for making HTTP requests. Simplifying the process to creating requests
import unittest
# Import for running and writing different test cases
```

```python
# Creating a request to the API to get a successful request
call = requests.get("https://api.artic.edu/api/v1/artworks")
print(call.status_code)
```

```python
class TestChicagoGalleryAPI(unittest.TestCase):
    def setUp(self):
        # URL for the Chicago Gallery API
        self.base_url = "https://api.artic.edu/api/v1/artworks"

    def test_api_status_code(self):
        # Test if the API returns a status code 200
        response = requests.get(self.base_url)
        self.assertEqual(response.status_code, 200, "API did not return status code 200.")

    def test_api_content_type(self):
        # Test if the API returns an image in the response
        response = requests.get(self.base_url)
        self.assertIn("image/", response.headers.get("Content-Type"), "API did not return an image content type.")

if __name__ == "__main__":
    unittest.TextTestRunner().run(unittest.TestLoader().loadTestsFromTestCase(TestChicagoGalleryAPI))
# Tests are successful if there is no failure message, just an "OK"
```
